```python
import os 
import pandas as pd
import numpy as np
import warnings

# 忽略所有警告
warnings.filterwarnings("ignore")
#碳排放监测数据所在dir
directory = r'C:\Users\RY\Desktop\火电碳排放影响因素识别与降碳措施研究\电厂调研数据\重要-五电厂碳排放数据汇总\宿州六号机组'
# List to hold each DataFrame
dataframes = []
# Loop through all files in the directory
for file in os.listdir(directory):
    #When an excel file is opened for example by MS excel a hidden temporary file is created in the same directory:~$神皖安庆一号机组23年1月.xlsx
    if not file.startswith("~") and file.endswith('.xlsx'):
        print(file)#打印该路径下读取到的文件名
        # Construct full file path
        file_path = os.path.join(directory, file)
        # Read the Excel file into a DataFrame
        df = pd.read_excel(file_path,skiprows=1)
        # Append the DataFrame to the list
        dataframes.append(df)
        #print(df)
# Concatenate all DataFrames into df for one year 
year_df_1_24 = pd.concat(dataframes, ignore_index=True)

# convert date from object type to datetime type 
year_df_1_24['时间'] = pd.to_datetime(year_df_1_24['时间'])
# set date to index
year_df_1_24.set_index('时间', inplace=True)
# sort data by date
year_df_1_24 = year_df_1_24.sort_index()
year_df_1_24['排放量绩效平均值(g/kwh)'] = pd.to_numeric(year_df_1_24['排放量绩效平均值(g/kwh)'], errors='coerce')
year_df_1_24_without0 = year_df_1_24[(year_df_1_24["有功功率平均值(MW)"] != 0)]
```

```python
#转换数据格式 obj——>float 以运算
year_df_1_24_without0['有功功率平均值(MW)'] = pd.to_numeric(year_df_1_24_without0['有功功率平均值(MW)'], errors='coerce')
year_df_1_24_without0['负荷率平均值(%)'] = pd.to_numeric(year_df_1_24_without0['负荷率平均值(%)'], errors='coerce')
year_df_1_24_without0['氧含量平均值(%)'] = pd.to_numeric(year_df_1_24_without0['氧含量平均值(%)'], errors='coerce')
```

```python
#确认计算涉及特征为数值型
year_df_1_24_without0.info()
```

```python
#根据公式计算每5min周期内的碳排放强度
year_df_1_24_without0["碳排放强度"] = ( (year_df_1_24_without0["C02浓度平均值(%)"]*44/22.4*10000)/1000*year_df_1_24_without0["烟气流量平均值(dNm3/h)"] ) / (1000 * year_df_1_24_without0["有功功率平均值(MW)"])
year_df_1_24_without0
```

```python
#计算值 碳排放强度 与导出值 排放量绩效平均值(g/kwh) 相差不大
year_df_1_24_without0.iloc[-20::]
```

```python

# resample to daily data
daily_1_24_data = year_df_1_24_without0.resample('D').agg({
    '有功功率平均值(MW)': 'mean',  # 计算每日平均负荷
    'C02浓度平均值(%)': 'mean',  # 计算每日二氧化碳体积浓度平均值
    '氧含量平均值(%)': 'mean',  # 计算每日平均氧含量
    '烟气流量平均值(dNm3/h)': 'mean',   # 计算每日平均烟气流量
    '碳排放量五分钟累计值(t)': 'sum',    # 计算每日碳排放量总和
    '碳排放强度': 'mean',   # 由5mins碳排放强度计算每日碳排放量均值
    '排放量绩效平均值(g/kwh)':'mean'
})
daily_1_24_data.iloc[-30::]
```
