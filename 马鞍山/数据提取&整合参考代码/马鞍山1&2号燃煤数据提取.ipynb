```python
import os
import pandas as pd
import numpy as np
# 马鞍山电厂燃料月度数据存放文件夹
fuel_directory = "../data/马鞍山电厂入炉煤质数据/马鞍山入炉煤统计表202308-202412"
#合并马鞍山电厂燃料月度数据 2308-2412月度文件为整年
# List to hold each DataFrame
dataframes = []
# Loop through all files in the directory
for file in os.listdir(fuel_directory):
    print(file)
    # Construct full file path
    file_path = os.path.join(fuel_directory, file)
    
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file_path,skiprows=6,skipfooter=2)
    
    # Append the DataFrame to the list
    dataframes.append(df)

# Concatenate all DataFrames into df for one year 
fuel_year_df = pd.concat(dataframes, ignore_index=True)
#重命名列名
fuel_year_df.columns = ["时间","总吨位（T)","全水分%","水分%","灰份%","挥发份%Vad","挥发份%Vdaf","全硫%ar","全硫%ad","高位热量","低位热量","固定碳%","#1煤量","#1热值","#2煤量","#2热值","飞灰%#1","飞灰%#2","炉渣%#1","炉渣%#2"]
# fuel_year_df.columns = ["时间","全水分%","灰份%","挥发份%Vad","低位热量","固定碳%"]	
# convert date from object type to datetime type 
fuel_year_df["时间"] = pd.to_datetime(fuel_year_df["时间"])
# set date to index
fuel_year_df.set_index("时间", inplace=True)
# sort data by date
fuel_year_df = fuel_year_df.sort_index()
#机组1&机组2整年日度数据
fuel_year_df = fuel_year_df.reset_index()
# 将整个 DataFrame 中的 "∕" 和 "/" 替换为 NaN
#fuel_year_df[(fuel_year_df["飞灰%#1"] =="∕")|(fuel_year_df["飞灰%#1"] =="/")]
fuel_year_df = fuel_year_df.replace(['∕', '/'], np.nan)
```
