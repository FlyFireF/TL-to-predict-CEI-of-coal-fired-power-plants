```python
import pandas as pd
import os
# 马鞍山电厂1号机组运行数据directory_1
directory_1 = "../data/马鞍山电厂运行数据/1号机组运行数据202308-202412"
# 马鞍山电厂2号机组运行数据directory_2
directory_2 = "../data/马鞍山电厂运行数据/2号机组运行数据202308-202412"
```

# 机组一运行数据读取

```python
dataframes_1 = []

for file in os.listdir(directory_1):
    print(file)
    file_path = os.path.join(directory_1, file)
    df=pd.read_excel(file_path,sheet_name='马鞍山电厂',skipfooter=11,header=1).iloc[4::,:-7]#跳过最后11行和最后7列
    cols=['项目','平均负荷','厂用电率','主汽压力', '主汽温度', '再热汽温', '给水温度', '排汽温度','排烟温度', '进风温度', '烟气含氧量']
    df=df[cols]
    dataframes_1.append(df)
    
df_all_1 = pd.concat(dataframes_1)

#更正部分列名使其更有意义
df_all_1 = df_all_1.rename(columns={'项目': '时间'})
# df_all_1 = df_all_1.rename(columns={'项目': '时间', '  \n真空度': '真空度'})

#将数据按照时间排序
# convert date from object type to datetime type  
df_all_1['时间'] = pd.to_datetime(df_all_1['时间'])
# set date to index
df_all_1.set_index('时间', inplace=True)
# sort data by date
df_all_1 = df_all_1.sort_index()
#reset index to column
df_all_1 = df_all_1.reset_index()

#添加负荷率列：平均负荷 / 机组容量（660MW) * 100
df_all_1["负荷率"] = df_all_1["平均负荷"] / 660 *100

#去除由于读取方式重复读取的日期8个 其排汽温度为空 
#从 2023 年 8 月 1 日到 2024 年 12 月 31 日一共是 519 天
df1 = df_all_1[df_all_1["排汽温度"].notnull()]
df1.to_excel("../已提取数据/马鞍山1号电厂运行数据202308-202412.xlsx",index=False)
```
