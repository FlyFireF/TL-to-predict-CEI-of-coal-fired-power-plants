```python
import pandas as pd
#读取 已提取运行数据
df_1_operation = pd.read_excel("../已提取数据/马鞍山1号电厂运行数据202308-202412.xlsx")
df_1_operation
```

```python
#读取 已提取燃煤煤质数据
df_1_fuel = pd.read_excel("../已提取数据/马鞍山1号煤质202308-202412.xlsx")
df_1_fuel
```

```python
#读取 已提取碳排放监测数据
df_1_emission = pd.read_excel("../已提取数据/马鞍山1号碳排放监测数据202308-202412.xlsx")
df_1_emission
```

```python
import functools as ft
#整合3种数据
data_frames = [df_1_operation,df_1_fuel,df_1_emission]
df_merged = ft.reduce(lambda  left,right: pd.merge(left,right,on=['时间'],
                                            how='outer'), data_frames)#.fillna(0)
df_merged
```

```python
df_merged.to_excel("../已整合数据/马鞍山1号数据整合202308-202412.xlsx")
```
