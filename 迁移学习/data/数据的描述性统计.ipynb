```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体 SimHei
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# data_aq12=pd.read_excel("安庆12.xlsx").drop('排汽温度', axis=1).drop('日期', axis=1)
# data_sz=pd.read_excel("宿州.xlsx").drop('日期', axis=1)
# data_mas=pd.read_excel("马鞍山.xlsx").drop('排汽温度', axis=1).drop('日期', axis=1)

data_aq12_ori=pd.read_excel("data_aq12_ori.xlsx").drop('排汽温度', axis=1).drop('日期', axis=1)
data_sz_ori=pd.read_excel("data_sz_ori.xlsx").drop('日期', axis=1)
data_mas_ori=pd.read_excel("data_mas_ori.xlsx").drop('排汽温度', axis=1).drop('日期', axis=1)
```

```python
data_aq12_ori.hist(bins=60, figsize=(15,9),color="red")
plt.tight_layout()
plt.show()
data_sz_ori.hist(bins=60, figsize=(15,9),color="red")
plt.tight_layout()
plt.show()
data_mas_ori.hist(bins=60, figsize=(15,9),color="red")
plt.tight_layout()
plt.show()
```

```python
def count_nan(df):
    # 检查是否有任何NaN值
    has_nan = df.isna().any(axis=1)
    # 检查CEI是否为0
    cei_zero = df['碳排放强度'] == 0
    # 组合条件
    missing_mask = has_nan | cei_zero
    missing_count = missing_mask.sum()
    print(missing_count)
count_nan(data_aq12_ori)
count_nan(data_sz_ori)
count_nan(data_mas_ori)
```
