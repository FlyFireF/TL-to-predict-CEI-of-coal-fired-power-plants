```python
import os
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.float_format', '{:.3f}'.format)
import warnings
warnings.filterwarnings("ignore")
# 设置中文字体为 SimHei（黑体）
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体 SimHei
plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题
```

```python
#读取 已提取运行数据
df_1_operation = pd.read_excel("../已提取数据/安庆1号运行数据2023.xlsx")
df_2_operation = pd.read_excel("../已提取数据/安庆2号运行数据2023.xlsx")
df_3_operation = pd.read_excel("../已提取数据/安庆3号运行数据2023.xlsx")
df_4_operation = pd.read_excel("../已提取数据/安庆4号运行数据2023.xlsx")

#统一时间列名 以合并
df_1_operation = df_1_operation.rename(columns = {"日期":"时间"})
df_2_operation = df_2_operation.rename(columns = {"日期":"时间"})
df_3_operation = df_3_operation.rename(columns = {"日期":"时间"})
df_4_operation = df_4_operation.rename(columns = {"日期":"时间"})


df_1_operation
#df_2_operation
#df_3_operation
#df_4_operation
```

```python
#读取 已提取燃煤煤质数据
df_1_fuel = pd.read_excel("../已提取数据/安庆1号煤质数据据2023.xlsx")
df_2_fuel = pd.read_excel("../已提取数据/安庆2号煤质数据据2023.xlsx")
df_3_fuel = pd.read_excel("../已提取数据/安庆3号煤质数据据2023.xlsx")
df_4_fuel = pd.read_excel("../已提取数据/安庆4号煤质数据据2023.xlsx")
df_1_fuel
#df_2_fuel
#df_3_fuel
#df_4_fuel
```

```python
#读取 已提取碳排放监测数据
df_1_emission = pd.read_excel("../已提取数据/安庆1号碳排放数据2023.xlsx")
df_2_emission = pd.read_excel("../已提取数据/安庆2号碳排放数据2023.xlsx")
df_3_emission = pd.read_excel("../已提取数据/安庆3号碳排放数据2023.xlsx")
df_4_emission = pd.read_excel("../已提取数据/安庆4号碳排放数据2023.xlsx")
df_1_emission
#df_2_emission
#df_3_emission
# df_4_emission
```

```python
import functools as ft
#整合3种数据
data_frames_1 = [df_1_operation,df_1_fuel,df_1_emission]
df_merged_1 = ft.reduce(lambda  left,right: pd.merge(left,right,on=['时间'],
                                            how='outer'), data_frames_1)#.fillna(0)

data_frames_2 = [df_2_operation,df_2_fuel,df_2_emission]
df_merged_2 = ft.reduce(lambda  left,right: pd.merge(left,right,on=['时间'],
                                            how='outer'), data_frames_2)#.fillna(0)

data_frames_3 = [df_3_operation,df_3_fuel,df_3_emission]
df_merged_3 = ft.reduce(lambda  left,right: pd.merge(left,right,on=['时间'],
                                            how='outer'), data_frames_3)#.fillna(0)

data_frames_4 = [df_4_operation,df_4_fuel,df_4_emission]
df_merged_4 = ft.reduce(lambda  left,right: pd.merge(left,right,on=['时间'],
                                            how='outer'), data_frames_4)#.fillna(0)
df_merged_2
```
